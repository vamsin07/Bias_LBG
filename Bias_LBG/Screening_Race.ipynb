```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Participant_data.csv')

columns_to_keep = ['last_screen_studyyr', 'num_screens', 'scr_days0', 'scr_days1', 'scr_days2', 'scr_group', 'race']
df = df[columns_to_keep]
print(df)


#print(value_counts)
```

```python
desired_column_1 = df['scr_group']
filtered_df = df[df['race'] != 1]
print(filtered_df)


```

```python
value_replacements = {
    1: 'Screen-detected cancer',
    2: 'Has nodule(s)',
    3: 'Screened, but no nodules',
    4: 'Never Screened',
    5: 'Other Lung cancer'
    # Add more replacements as needed
}

df_1 = filtered_df['scr_group'].replace(value_replacements)
print(df_1)
```

```python
value_replacements_race = {
    1: 'White',
    2: 'Black or African-American',
    3: 'Asian',
    4: 'American Indian or Alaskan Native',
    5: 'Native Hawaiian or Other Pacific Islander',
    6: 'More than one race',
    7: 'Participant refused to answer',
    95: 'Missing data form - form is not expected to ever be completed',
    96: 'Missing - no response',
    98: 'Missing - form was submitted and the answer was left blank',
    99: 'Unknown/ decline to answer'
}
df_2 = filtered_df['race'].replace(value_replacements_race)
print(df_2)
```

```python
joined_df = pd.concat([df_1, df_2], axis=1)
joined_df.columns = ['Screening_Group', 'Race']
print(joined_df)
```

```python
not_screened = joined_df[joined_df['Screening_Group'] == 'Never Screened']
print(not_screened)
```

```python
non_screened_counts = not_screened['Race'].value_counts()
non_screened_counts.plot(kind='bar')
plt.title('Distribution of Non-Screened Individuals by Race')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()
plt.savefig('Distribution of Non_screened.png')
```
